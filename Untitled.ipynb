```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time
from selenium.webdriver.common.by import By

# *** VARIAVEIS
strUrl = 'https://www.google.com'
strTextToSearch = "nike"


# *** DRIVERS
driver = webdriver.Chrome()
# *** PASSO 01: Abre o navegador e carrega a URL 
driver.get(strUrl)

# *** PASSO 02: Verifica se o elemento de busca foi encontrado com sucesso ou ocorreu algum erro.
isOk = driver.find_element_by_name('q')

if isOk:
    print('PASSO 01: O campo de pesquisa foi encontrado: SUCESSO')
else:
    print('PASSO 01: O campo de pesquisa foi encontrado: ERRO')


# *** PASSO 03: Pesquisa o o nome inserido na variavel strTextToSearch
driver.find_element_by_name('q').send_keys(strTextToSearch + Keys.ENTER)

validacaoDeBusca = driver.find_elements_by_xpath('//*[@id="voiceSearchButton"]')

if validacaoDeBusca:
    print('Busca não encontrada')
else:
    print('Busca encontrada com sucesso')
 

# *** PASSO 04: Endereço de um elemento HTML e clique
driver.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div/div/div/div/div/div[1]/a/h3').click()


time.sleep(1)
driver.close()

# *** VARIAVEIS
strUrl1 = 'https://www.google.com'
strTextToSearch1 = "nike"


# *** DRIVERS
driver1 = webdriver.Firefox()
# *** PASSO 01: Abre o navegador e carrega a URL 
driver1.get(strUrl)

# *** PASSO 02: Verifica se o elemento de busca foi encontrado com sucesso ou ocorreu algum erro.
isOk1 = driver1.find_element_by_name('q')

if isOk1:
    print('PASSO 01: O campo de pesquisa foi encontrado: SUCESSO')
else:
    print('PASSO 01: O campo de pesquisa foi encontrado: ERRO')


# *** PASSO 03: Pesquisa o o nome inserido na variavel strTextToSearch
driver1.find_element_by_name('q').send_keys(strTextToSearch1 + Keys.ENTER)

validacaoDeBusca1 = driver1.find_elements_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[1]/div/span/svg')

if validacaoDeBusca1:
    print('Busca não encontrada')
else:
    print('Busca encontrada com sucesso')
 
time.sleep(1)

# *** PASSO 04: Endereço de um elemento HTML e clique
driver1.find_element_by_xpath('/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div/div/div/div[1]/a/div/cite').click()


time.sleep(1)
driver1.close()
```
